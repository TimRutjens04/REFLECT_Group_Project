# Grounding DINO vs closed-vocabulary detectors

This notebook compares the open-vocabulary reference model `Grounding DINO Base` against three efficient closed-vocabulary detectors on the 24 images in `data/`.

Shared evaluation uses manual frame-level presence labels for `apple`, `cup`, and `person`.

## Model choices

- `YOLO11n`: latest lightweight Ultralytics detection model on COCO-80.
- `RT-DETRv2-R18`: compact real-time transformer detector.
- `Faster R-CNN ResNet50 FPN V2`: strong Torchvision closed-vocabulary baseline.
- `Grounding DINO Base`: open-vocabulary reference.

All choices were kept lightweight enough to stay practical on CPU-only hardware.
